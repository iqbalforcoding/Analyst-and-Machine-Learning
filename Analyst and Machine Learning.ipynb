## **Import Library**

Kita membutuhkan beberapa library yaitu pandas, datetime dan juga matplotlib sebagai alat pendukungnya.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
```

Kita lakukan pembacaan data csv yang telah kita upload dan ditaruh pada sample_data

```python
df = pd.read_csv('sample_data/ecommerce_sales.csv')
df
```

Jadi, apabila isi dari df ada yang sama dengan nama kolomnya. Otomatis akan tidak ditampilkan. Yang hanya ditampilkan yaitu data yang tidak sama dengan nama kolom tabel.

```python
df = df[(df != df.columns).all(axis=1)]
df
```

Setelah itu, apabila ada isi yang NaN. Kita akan menghapusnya agar data tersebut menjadi bagus.

```python
df = df.dropna()
df
```

Di sini, pada kolom Quantity Ordered dan Price Each tipe datanya akan diubah menjadi float. Agar dapat menghitung Quantity Ordered dikali Price Each yang menghasilkan kolom T_Price.

```python
df['Quantity Ordered'] = df['Quantity Ordered'].astype(float)
df['Price Each'] = df['Price Each'].astype(float)
df.dtypes
```

Kita membuat sebuah kolom baru yang dimana isinya didapat dari kolom Quantity Ordered	dikali Price Each.

```python
df['T_Price'] = df['Quantity Ordered']*df['Price Each']
df
```

Di sini kita membuat bulan terlebih dahulu. Agar memudahkan melihat pada bulan apa penjualan paling tinggi dan rendah.

```python
df['Month'] = df['Order Date'].str[0:2]
df['Month'] = df['Month'].astype('int')
df
```

**Sorting data perbulan**

Karena data diawal masih sangat abstrak. Maka kita akan membuatnya berdasarkan perbulan. Dan transaksi yang terjadi dalam satu bulan akan ditambahkan menjadi satu. Lalu kita akan melihat data yang akan ditampilkan selama 1 tahun

```python
satu_bulan= df['T_Price'].groupby(df['Month'])
satu_bulan = satu_bulan.sum()
satu_bulan
```

Maka, total belanja paling tinggi senilai 4613443.34 yang terjadi pada bulan 12.

```python
satu_bulan.max()
```

Maka, total belanja paling tinggi senilai 1822256.73 yang terjadi pada bulan 1.

```python
satu_bulan.min()
```

# **Visualisasi Data**
Setelah melihat visualisasi di bawah ini dapat ditarik kesimpulan. Transaksi besar yang terjadi dalam satu bulan yaitu pada bulan 12 atau Desember. Dan juga transaksi terkecil terjadi pada bulan 1 atau Januari. Oleh karena itu, pada saat akhir tahun sangat cocok apabila perusahaan ingin menaikkan jumlah barang.

```python
plt.figure(figsize=(5,5))
satu_bulan.plot(kind='bar',title='Month',legend=True,color='r')
plt.ylabel('Total Price')
plt.xlabel('Month')
plt.show()
```

**Product Terlaris dan Tidak Terlaris**

Kita membuat terlebih dahulu sebuah product agar dapat mensortir dengan product. Kemudian, dalam satu product itu kita jumlahkan agar dapat melihat bahwa jumlah pembelian dalam product tersebut.

```python
product = df.groupby('Product')
q_order = product.sum()['Quantity Ordered']
q_order
```

# **Visualisasi Data**
Kita visualisasikan ke dalam bentuk chart. Dengan menggunakan plot kita membuat sebuah chart bar dengan title Product Terlaris. Di sini kita dapat melihat bahwa product AAA Batteries (4-pack) merupakan yang terlaris diantara semuanya. Dan product LG Dryer merupakan product yang kurang diminati diantara semuanya. Maka, kalau bisa product AAA Batteries (4-pack) dapat diperbanyak stoknya. 

```python
q_order.plot(kind='bar',title='Product terlaris',legend=True,color='r')
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.show()
```

# **Sortir pemesanan berdasarkan jam, dan menit**
Lalu, kita akan mensortir pemesanan berdasarkan jam dan menitnya. Agar kita dapat mengetahui waktu yang tepat ketika ingin memasarkan produk. Di sini kita buat kolom Hour terlebih dahulu dengan mengambil data dari order date. Dengan cara memotong string yang ada.

```python
df['Hour'] = df['Order Date'].str[9:11]
df['Hour'] = df['Hour'].astype('int')
df
```

Lalu, kita buat juga kolom Minute dengan mengambil data dari kolom order date. Dengan cara memotong string yang ada.

```python
df['Minute'] = df['Order Date'].str[12:14]
df['Minute'] = df['Minute'].astype('int')
df
```

Disini, kita buat beberapa array yang akan digunakan ketika memvisualisasi.

```python
a = np.arange(0,24)
b = np.arange(0,60)
```

Di sini, kita buat berdasarkan jam terlebih dahulu agar memudahkan nanti. Dapat dilihat jumlah pemesan paling sedikit dapat terlihat pada jam 3. Dan pemesanan terbesar terlihat pada jam 19.

```python
jam1 = df.groupby('Hour')
jam1 = jam1.sum()['Quantity Ordered']
jam1
```

Di sini, kita buat berdasarkan jam terlebih dahulu agar memudahkan nanti. Dapat dilihat jumlah pemesan paling sedikit dapat terlihat pada menit 6. Dan pemesanan terbesar terlihat pada menit 48.

```python
menit1 = df.groupby('Minute')
menit1 = menit1.sum()['Quantity Ordered']
menit1
```

# **Visualisasi Data**
Dengan menjadikan sebuah chart, kita dapat melihatnya dengan jelas pada jam-jam berapa mengiklankan agar para pembeli meningkat. Ketika siang hari dan juga malam hari.

```python
jam1.plot(kind='line',title='Product terlaris',legend=True,color='m')
plt.ylabel('Ordered')
plt.xticks(a)
plt.xlabel('Hour')
plt.grid()
plt.show()
```

```python
menit1.plot(kind='line',title='Product terlaris',legend=True,color='m')
plt.ylabel('Ordered')
plt.xticks(b)
plt.xlabel('Minute')
plt.grid()
plt.show()
```
